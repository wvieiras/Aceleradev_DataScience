```python
import pandas as pd
```

```python
data = pd.read_csv('desafio1.csv')
```

```python
data.head(10)
```

```python
data['estado_residencia'].value_counts()
```

```python
data.shape
```

```python
resume = data.groupby(['estado_residencia'])['pontuacao_credito'].agg([pd.Series.mode,'median','mean','std'])
resume.reindex()
```

```python
resume.rename(columns={'mode': 'moda', 'median': 'mediana', 'mean': 'media', 'std': 'desvio_padrao'}, inplace = True)
resume.sort_index(ascending = False)
```

```python
resume = resume.T
```

```python
columnsTitles = ['SC','RS', 'PR']
resume = resume.reindex(columns=columnsTitles)
```
